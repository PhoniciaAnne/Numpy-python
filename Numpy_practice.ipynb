```python
static_array = [[np.random.uniform(0,1) for i in range(1536)] for _ in range(10)]
print(static_array)
```

```python
def get_dot_product(arr1,arr2):
  return np.dot(arr1,arr2)

def get_cosine_similarity(arr1, arr2):
    dot_product = np.dot(arr1, arr2)
    norm_arr1 = np.linalg.norm(arr1)
    norm_arr2 = np.linalg.norm(arr2)
    cosine_similarity = dot_product / (norm_arr1 * norm_arr2)
    return cosine_similarity
```

```python
for test in range(10):
  test_example = [np.random.uniform(0,1) for _ in range(1536)]

  dot_product_results = []
  cosine_similarity_results = []
  for trained_example in static_array:
    dot_product = get_dot_product(test_example,trained_example)
    cosine_similarity = get_cosine_similarity(test_example,trained_example)
    dot_product_results.append(dot_product)
    cosine_similarity_results.append(cosine_similarity)
  print("MAximum cosine Similarity",np.max(cosine_similarity_results))
  print("MAximum dot product",np.max(dot_product_results))
```
